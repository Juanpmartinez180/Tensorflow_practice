```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
init_found = 4000
gain_margin = 0.15  #1 = 100%
invested_monts = 15

money = [init_found]
monts = [0]
montly_revenue = [0]
for i in range(invested_monts):
    temp_gain = money[i]+money[i]*gain_margin
    money.append(temp_gain)
    monts.append(i+1)
    montly_revenue.append(money[i]*gain_margin)
```

```python
plt.plot(monts, money)
plt.title('Total Investment over Time')
plt.xlabel('Monts')
plt.ylabel('$USD')
plt.show()
```

```python
plt.plot(monts, montly_revenue)
plt.title('Montly revenue')
plt.xlabel('Monts')
plt.ylabel('USD$')
plt.show()
```
