# Cholesky matrix decomposition 

- The idea is to decompose the original matrix (A) in to a 2 smaller matrix. This method bring us more computational efficiency and therefore less runinng time to solve problems like linear regression. Where:

$$A = LU$$
- $L$: Upper triangular matrix
- $U$: Lower triangular matrix
- $L^T = U$

Then to solve the linear regression problem..

$$ Ax = b $$ or
$$A^TAx = A^Tb$$
$$LUx = b$$

In this form we can try to solve the equation but in two steps:

$$Ly = b$$
$$Ux = y$$

```python
%config Completer.use_jedi = False
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
session = tf.Session()
```

```python
x_vals = np.linspace(0,10,100)
y_vals = x_vals + np.random.normal(0,1,100)
plt.plot(x_vals, y_vals, 'o', label = 'Data')
```

```python
x_vals_column = np.transpose(np.matrix(x_vals)) 
ones_column = np.transpose(np.matrix(np.repeat(1,100)))
A = np.column_stack((x_vals_column, ones_column))
b = np.transpose(np.matrix(y_vals))
```

```python
A_tensor = tf.constant(A)
b_tensor = tf.constant(b)
```

```python
tA_A = tf.matmul(tf.transpose(A_tensor), A_tensor)   #Inverse matrix solver
L = tf.cholesky(tA_A)
tA_b = tf.matmul(tf.transpose(A_tensor), b_tensor)
y_sol = tf.matrix_solve(L, tA_b)
x_sol = tf.matrix_solve(tf.transpose(L), y_sol)

sol_eval = session.run(x_sol)
```

```python
slope = sol_eval[0][0]
intercept = sol_eval[1][0]
```

```python
print('Regression equation using Cholesky decomposition = ', str(np.round(slope,3))+ ' x + '+ str(np.round(intercept,3))   )
```

```python
best_fit = []
for i in x_vals:
    best_fit.append(slope*i + intercept)
    
plt.plot(x_vals, y_vals, 'bo', label = 'Original data')
plt.plot(x_vals, best_fit, 'r-', label = 'LU decomp. linear regression', linewidth = 3)
plt.legend(loc = 'upper left')
plt.show()
```
